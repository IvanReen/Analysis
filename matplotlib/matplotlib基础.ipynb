### X、Y轴坐标刻度
plt.xticks()和plt.yticks()方法  

- 需指定刻度值和刻度名称  plt.xticks([刻度列表],[名称列表])
- 还可以传入fontsize、rotation、color等参数设置

````python
x = np.linspace(-np.pi,np.pi,100)
plt.plot(x,np.sin(x))
# plt.xticks([-10,0,10])  # 以列表的形式传入刻度 注意：刻度的值不要超过取值的范围太多
# plt.xticks([-4,0,4])
# plt.xticks([-3.14,-3.14/2,0,3.14/2,3.14])
plt.xticks([-3.14,-3.14/2,0,3.14/2,3.14],['-$\pi$','-$\pi$/2',0,'$\pi$/2','$\pi$'])  # 还可以在后面 以列表的形式传入别名
plt.xticks([-3.14,-3.14/2,0,3.14/2,3.14],['-$\pi$','-$\pi$/2',0,'$\pi$/2','$\pi$'],fontsize=20,color='red',rotation=20) 
````

````python
x = np.linspace(-np.pi,np.pi,100)
axes = plt.subplot()
axes.plot(x,np.sin(x))
axes.set_xticks([-3.14,-3.14/2,0,3.14/2,3.14])
# set_xticks只是用来设置轴线上的刻度 其他的不用这个方法来设置
axes.set_xticks([-3.14,-3.14/2,0,3.14/2,3.14],['-$\pi$','-$\pi$/2',0,'$\pi$/2','$\pi$'])  # 传入别名也不会出错 但是不会有效果
axes.set_xticklabels(['-$\pi$','-$\pi$/2',0,'$\pi$/2','$\pi$'],fontsize=20,color='red',rotation=20)
````

## 三、2D图形

### 直方图

【直方图的参数只有一个x！！！不像条形图需要传入x,y】

hist()的参数
+ bins  
    - 可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
+ density (normed) 
    - 如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
+ color  
    - 指定直方图颜色。可以是单一色值或颜色序列。如果指定了多个数据集合，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
+ orientation  
    - 通过设置orientation为horizontal创建水平直方图。默认值为vertical

````python
data = np.random.randint(0,100,size=100)
data
````

````python
# df = DataFrame(data)
# df.plot(kind='hist')
# 频率分布直方图 
# 参数 x是传入的数据 bins用来指定将整体分为几个区间默认是10 density参数False显示频率True显示概率
# 返回值 1各个区间有多少样本 2区间的范围
# plt.hist(data,bins=2)
plt.hist(data,bins=5,density=True,color='red',orientation='horizontal')
````

### 条形图

【条形图有两个参数x,y】

- width 水平方向的长度
- height 竖直方向的长度

bar()、barh()

data = np.array([1,4,7,2,8,5])

index = [1,2,3,4,5,6]

````python
data = np.array([1,4,7,2,8,5,9])
index = [1,2,3,4,5,6,7]
# 参数 x是索引 height是数值 这两个参数必须传
# plt.bar(x=index,height=data)
plt.bar(x=index,height=data,width=0.5)  # width用来设置条的宽度 按照比例设置 1是满的
# height 竖直方向的长度 width 水平方向的长度
````

````python
data = np.array([1,4,7,2,8,5,9])
index = [1,2,3,4,5,6,7]
# 参数 y是索引 height是竖直方向的长度 width是水平方向的长度
plt.barh(y=index,height=0.5,width=data)
````

````python
data = np.array([1,4,7,2,8,5])
index = [1,2,3,4,5,6]
plt.axes(polar=True)  # 开启极坐标
plt.bar(x=index,height=data) 
````

### 饼图

【饼图也只有一个参数x！】

pie()
饼图适合展示各部分占总体的比例，条形图适合比较各部分的大小

普通各部分占满饼图

````python
data = np.array([30,50,70,90])  # 所有数值的和大于1 绘制的是各部分的占比
plt.pie(data)
````

普通未占满饼图

````python
data = np.array([0.1,0.2,0.3,0.2])  # 如果和不到1 就会有缺口
plt.pie(data)
````

饼图阴影、分裂等属性设置
- labels参数设置每一块的标签；
- labeldistance参数设置标签距离圆心的距离（比例值,只能设置一个浮点小数）
- autopct参数设置比例值的显示格式(%1.1f%%)；
- pctdistance参数设置比例值文字距离圆心的距离
- explode参数设置每一块顶点距圆形的长度（比例值,列表）；
- colors参数设置每一块的颜色（列表）；
- shadow参数为布尔值，设置是否绘制阴影
- startangle参数设置饼图起始角度

````python
data = np.array([30,50,70,90])  # 所有数值的和大于1 绘制的是各部分的占比
plt.pie(data,
        labels=['1','2','3','4'],
        labeldistance=0.5,
        autopct='%1.1f%%',
        pctdistance=0.8,
        explode=[0.2,0,0,0],
        colors=['red','green','blue','orange'],
        shadow=True,
        startangle=90
       )
````

## 四、图形内的文字、注释、箭头

控制文字属性的方法:  

| pyplot函数     | API方法                | 描述                     |
| :-------------: |:------------------------------:| :---------------------------------:|
| plt.suptitle()     | mpl.figure.Figure.suptitle() |  为Figure对象添加中心化的标题     |
| plt.figtext()     |  mpl.figure.Figure.text()    |  在Figure对象的任意位置添加文字    |
| axes.set_title()       | mpl.axes.Axes.set_title()   |  为Axes对象添加标题            |
| axes.text()       |  mpl.axes.Axes.text()       | 在Axes对象的任意位置添加文字     |
| xlabel()      |  mpl.axes.Axes.set_xlabel()   | 为X轴添加标签               |
| ylabel()      |  mpl.axes.Axes.set_ylabel()   | 为Y轴添加标签               |

所有的方法会返回一个matplotlib.text.Text对象

### 图形内的文字
text()  
![](14.png)

````python
axes = plt.subplot()
plt.suptitle('suptitle')  # 整个图的标题
# figtext必须传入 x y 来设置文本位置
plt.figtext(0.5,0.5,'figtext')  # 整个图中按照位置设置文字
axes.set_title('axes title')  # 坐标系的标题
# x, y, s,  fontdict=None
axes.text(0.2,0.8,'axes text')
````

### 注释
annotate()
- xy参数设置箭头指示的位置
- xytext参数设置注释文字的位置  
- arrowprops参数以字典的形式设置箭头的样式  
    - width参数设置箭头长方形部分的宽度
    - headlength参数设置箭头尖端的长度，  
    - headwidth参数设置箭头尖端底部的宽度
    - shrink参数设置箭头顶点、尾部与指示点、注释文字的距离（比例值），可以理解为控制箭头的长度

````python
# 留着
data = np.random.randint(0,10,size=10)
index = np.arange(0,10,1)
data
index
````

````python
plt.plot(index,data)
# plt.annotate(s='this is very important',xy=[5,5],fontsize=20,color='red',rotation=20)
# plt.annotate(s='this is very important',xy=[6,4],xytext=[5,5],arrowprops={'width':10})
# plt.annotate(s='this is very important',xy=[6,4],xytext=[5,5],arrowprops={'width':2,'headlength':5,'headwidth':10})
plt.annotate(s='this is very important',xy=[6,4],xytext=[5,5],arrowprops={'arrowstyle':'->'})
````

    如下都是arrowstyle可以选择的风格样式
    
    ``'->'``       head_length=0.4,head_width=0.2
    ``'-['``       widthB=1.0,lengthB=0.2,angleB=None
    ``'|-|'``      widthA=1.0,widthB=1.0
    ``'-|>'``      head_length=0.4,head_width=0.2
    ``'<-'``       head_length=0.4,head_width=0.2
    ``'<->'``      head_length=0.4,head_width=0.2
    ``'<|-'``      head_length=0.4,head_width=0.2
    ``'<|-|>'``    head_length=0.4,head_width=0.2
    ``'fancy'``    head_length=0.4,head_width=0.4,tail_width=0.4
    ``'simple'``   head_length=0.5,head_width=0.5,tail_width=0.2
    ``'wedge'``    tail_width=0.3,shrink_factor=0.5

````python
plt.figure(figsize=(10,6))
axes = plt.subplot(projection='3d')
# axes.plot_surface(X,Y,Z)
p = axes.plot_surface(X,Y,Z,cmap='rainbow')
plt.colorbar(p,shrink=0.5)
````
