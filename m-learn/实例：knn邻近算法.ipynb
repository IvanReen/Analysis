### 2. 预测年收入是否大于50K美元

读取adults.csv文件，最后一列是年收入，并使用KNN算法训练模型，然后使用模型预测一个人的年收入是否大于50

```python
df = pd.read_csv('./data/adults.csv')
df.head()
```

获取年龄、教育程度、职位、每周工作时间作为机器学习数据  
获取薪水作为对应结果

```python
data = df.loc[:,['age','education','occupation','hours_per_week']]  # 特征
```

```python
target = df['salary']
```

```python
data
```

数据转换，将String类型数据转换为int

【知识点】map方法，进行数据转换

```python
# 给值是字符串的那些列 变成 对应的数值 这样 机器学习模型才能计算
```

```python
levels = data['education'].unique()
levels
```

```python
levels=='Bachelors'
np.argwhere(levels=='Bachelors')  # np.argwhere 传入条件 就可以找到 符合条件的值的位置
np.argwhere(levels=='Masters')  # 在levels里面 每一个 受教育程度的名字 都可以对应一个编号
# 这样我们就可以把 字符串 变成 数值了
np.argwhere(levels=='Masters')[0,0]  # 这个是 Master的编号
```

```python
# 把 字符串 变成 整数
def str2int(item):
    # item是教育程度中的每一项 比如 'Bachelors'  'Masters'
    # 通过np.argwhere找到这一项 在levels中的索引号 这个索引号就是编号 把找到的编号返回即可
    return np.argwhere(levels==item)[0,0]
```

```python
data.education
```

```python
data.education = data.education.map(str2int)  # map 映射 可以把一个值映射成另一个值
```

```python
data.dtypes
```

```python
# 职位 的 字符串 也要变成 对应的数值
```

```python
# 封装函数 传入df和要转换的列的名字 就把这个列的字符串变成数值
def transform_dtype(df,column_name):
    # 看看传入的df中指定的这一列 有哪些独特的值
    u = df[column_name].unique()
    def str2int(item):
        return np.argwhere(u==item)[0,0]
    # 传入的df的指定列 使用map做映射
    df[column_name] = df[column_name].map(str2int)
```

```python
transform_dtype(data,'occupation')
```

```python
data.dtypes
```

```python
data
```

切片：训练数据和预测数据

```python
# 把处理好的数据 分成 训练集 和 测试集
X_train = data[:30000]
y_train = target[:30000]
X_test = data[30000:]
y_test = target[30000:]
```

生成算法

```python
knn = KNeighborsClassifier()
```

第一步：训练数据

```python
knn.fit(X_train,y_train)
```

第二步：预测数据

```python
y_ =knn.predict(X_test)
y_
```

```python
knn.score(X_test,y_test)
```

### 3. 癌症预测

```python
cancer = pd.read_table('./data/cancer.tsv')
cancer.head()
```

```python
data = cancer.iloc[:,2:]  # 所有样本的特征列
target = cancer.iloc[:,1]  # 所有样本的标签
# target目标值中，M恶性，B良性，没事儿
display(data.head(),target.head())
```

```python
data.shape
```

```python
# 打乱数据的顺序 并分成 训练集 和 测试集
```

```python
# train_test_split 这个函数直接就可以把 数据集 分成 训练集和测试集 而且还会打乱顺序
from sklearn.model_selection import train_test_split
```

```python
#  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
# train_test_split函数 参数： X 特征数据集 y 结果集合 test_size测试集占的比例 random_state对随机数的设定
# 返回值 分别是X_train, X_test, y_train, y_test这样的顺序 （一定要注意这个顺序）
X_train, X_test, y_train, y_test = train_test_split(data,target,test_size=0.1,random_state=42)
```

```python
X_train
```

```python
# 获取模型并训练
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
y_
```

```python
y_test
```

```python
knn.score(X_test,y_test)
```

```python
pd.crosstab(y_,y_test,rownames=['predict'],colnames=['real'])
# M恶性 B良性
```

```python
data
```

```python
# 归一化 把各个范围的值 统一成 0到1之间的值
# 最大值-最小值 是 范围
# 当前值-最小值 是 当前值的相对大小
# (当前值-最小值)/(最大值-最小值)  0到1之间的 相对大小的值
```

```python
# data中每一列特征都要做归一化处理
# data中的每一列
data.columns
```

```python
for cols in data.columns:
    # print(cols)
    # data[cols] 是 每一列数据
    data_max = data[cols].max()
    data_min = data[cols].min()
    # data[cols]-data_min 这个Series里面的每一个值都减去data_min
    # 每一个值都这样计算 然后在重新赋值给这一列
    data[cols] = (data[cols]-data_min)/(data_max-data_min)
```

```python
data
```

```python
# 从完整的数据集中 分割出 训练集和测试集
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(data,target,test_size=0.1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
y_
```

```python
y_test
```

```python
knn.score(X_test,y_test)
```

```python
pd.crosstab(y_,y_test,rownames=['predict'])
```
